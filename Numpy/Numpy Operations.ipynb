# NumPy Operations

## Arithmetic

You can easily perform array with array arithmetic, or scalar with array arithmetic. Let's see some examples:

```python
import numpy as np
arr = np.arange(0,10)
```

```python
arr + arr
```

```python
arr * arr
```

```python
arr - arr
```

```python
arr**3
```

## Universal Array Functions



```python
#Taking Square Roots
np.sqrt(arr)
```

```python
#Calcualting exponential (e^)
np.exp(arr)
```

```python
np.max(arr) #same as arr.max()
```

```python
np.sin(arr)
```

```python
np.log(arr)
```
